# Importing The Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection & Data Preprocessing

```python
path_breastCancer = '/content/drive/MyDrive/Projects/data.csv'
```

```python
df = pd.read_csv(path_breastCancer)
```

```python
# Gives you first five rows of data
df.head()
```

```python
#Check the shape of the dataframe
df.shape
```

```python
# Check the Information of dataframe
df.info()
```

```python
# Checking the missing value
df.isnull().sum()
```

```python
del df['Unnamed: 32']
```

```python
#Drop the column has Nan value
df.drop(['id'],axis=1,inplace=True)
```

```python
df.info()
```

```python
# Stastical measure about data
df.describe()
```

```python
# enumerate the diagnosis column such that M = 1, B = 0
df['diagnosis'] = df['diagnosis'].apply(lambda x: '1' if x == 'M' else '0')
```

```python
df.head()
```

```python
# Checking the distribution of target variable
print(df.groupby('diagnosis').size())
```

1-- M

0-- B

```python
df.groupby('diagnosis').mean()
```

Separating the Features and Target

```python
X = df.drop(columns='diagnosis',axis = 1)
Y = df['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data And Testing Data

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size=0.2,random_state =2)
```

```python
print(X.shape,X_Train.shape,X_Test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_Train,Y_Train)
```

Model Evaluation

Accuracy Score

```python
# Accuracy on the training data
X_Train_Prediction = model.predict(X_Train)
Training_data_Prediction = accuracy_score(Y_Train,X_Train_Prediction)
```

```python
print('Accuracy on Training data =',Training_data_Prediction )
```

```python
# Accuracy on the test data
X_Test_Prediction = model.predict(X_Test)
Test_data_Prediction = accuracy_score(Y_Test,X_Test_Prediction)
```

```python
print('Accuracy on Test data =',Test_data_Prediction )
```

# Building a predictive system

```python
input_data = (12.46,24.04,83.97,	475.9,	0.1186,	0.2396,	0.2273,0.08543,	0.203,	0.08243,	0.2976,	1.599,	2.039	,23.94,0.007149,	0.07217,0.07743,0.01432,	0.01789	,0.01008,15.09,	40.68,97.65,711.4	,0.1853,	1.058,1.105,0.221,	0.4366,	0.2075)

# change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for the one datapoint
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshape)

if (prediction[0]==0):
  print("Breast Cancer is B")
else:
  print("Breast Cancer is M")
```
